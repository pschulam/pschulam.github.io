# Diffusion Processes

This post is an introduction to the idea of a diffusion process, which is a rich family of probabilistic models in *continuous time*. I'll describe how a diffusion process is defined and give a brief overview of the steps we take to derive probability density functions for the process.

We're going to work with the simplest type of diffusion process: linear, time homogeneous diffusion processes. These processes are described using notation that looks like a differential equation, but that has a slightly different meaning. Here is our simple diffusion process:

$$
{\sf d} x = F x {\sf d}t + G u {\sf d}t + L w {\sf d}t.
$$

The variable $x \in \mathbb{R}^p$ denotes the *state* of the process, $u \in \mathbb{R}^q$ is the *input*, and $w \in \mathbb{R}^r$ is white noise. All three of these variables are functions of time, so to be more explicit we would write $x(t)$, $u(t)$, and $w(t)$.

The first thing to note here is that the expression above *looks* like a differential equation, but it isn't because we cannot algebraically divide by ${\sf d}t$ and get something meaningful. If you're familiar with some of the basic properties of white noise, you might remember that it is nowhere differentiable, which means that $\frac{{\sf d} w}{{\sf d}t}$ does not exist. I'm glossing over details, but at a high-level this is why we can't treat the expression above like a differential equation.

## Itô Stochastic Integrals

So if it's not a differential equation, then what is it? It's a shorthand for the following integral equation:

$$
x(t) - x(t_0) =
\int_{t_0}^t F x {\sf d}t +
\int_{t_0}^t G u {\sf d}t +
\int_{t_0}^t L w {\sf d}t.
$$

The first two integrals are standard Riemann (or Lebesgue) integrals, but the third violates conditions needed to define an integral. In particular, the white noise process $w$ is unbounded and discontinuous. The Itô stochastic integral gives precise meaning to this expression.

To define the Itô integral, we'll use the idea of Brownian motion $\beta(t)$, one of the simplest stochastic processes. The key property that we'll use is that white noise can be defined as the formal derivative of Brownian motion. So we can write

$$
\frac{{\sf d} \beta}{{\sf d}t} = w
\implies
{\sf d} \beta = w {\sf d}t.
$$

In the third integral above, we can therefore make an algebraic substitution of ${\sf d}\beta$ for $w {\sf d}t$. Once we make this substitution, then we define the Itô integral to be

$$
\int_{t_0}^t L {\sf d} \beta =
\lim_{n \to \infty}
\sum_{k} L [ \beta(t_{k+1}) - \beta(t_k) ].
$$

The increment between two time points of Brownian motion is Gaussian with scale $\sigma = t_{k+1} - t_k$, so for a finite grid $t_0, \ldots, t_K$ the right-hand expression is a sum of Gaussian distributions and has a meaning. As we let the grid become increasingly fine, we get a sequence of random variables, which we can reason about rigorously using probability theory. We've made progress on better understanding the way that we express a diffusion process, but still don't know how to do anything useful with it (like compute probabilities). We'll cover that next.

## Itô Formula

The standard technique for solving linear differential equations is to use an *integrating factor*. Suppose we have the simple system

$$
{\sf d}x = \lambda x {\sf d}t,
$$

where $x(t) \in \mathbb{R}$ and $\lambda \in \mathbb{R}$ is a known parameter. To solve for $x(t)$ with initial condition $x(0)$ given, we multiply both sides by $\exp\{-\lambda t\}$ (this is our integrating factor). Rearranging, and dividing by ${\sf d}t$ we have

$$
\exp\{-\lambda t\} \frac{{\sf d}x}{{\sf d} t} - \exp\{-\lambda t\} \lambda x = 0.
$$

Now, using the chain rule we see that

$$
\frac{{\sf d} \exp\{-\lambda t\} x}{{\sf d}t} =
\exp\{-\lambda t\}\frac{{\sf d}x}{{\sf d}t} - \exp\{-\lambda t\}\lambda x.
$$

Substituting and integrating from $0$ to $t$, we have that

$$
\exp\{-\lambda t\} x(t) - x(0) = 0
\implies
x(t) = \exp\{\lambda t\} x(0).
$$

Unfortunately, we can't, in general, use this technique to solve stochastic differential equations. Recall that we cannot divide our stochastic differential equation by ${\sf d}t$, which was a key step in the integrating factor method. Using Itô's formula, however, we can do something analagous.

Itô's formula gives use the stochastic version of the chain rule. At a high-level, if $x(t)$ is a diffusion process and $\phi(x, t)$ is a function of the process, then Itô's formula shows us how to write down the stochastic differential equation for the derived process $\phi(x, t)$. I'll simply present the result without proof to keep things simple:

$$
{\sf d}\phi =
\frac{\partial \phi}{\partial t} {\sf d}t +
\sum_i \frac{\partial \phi}{\partial x_i} {\sf d}x_i +
\frac{1}{2} \sum_{i, j} \Big( \frac{\partial^2 \phi}{\partial x_i \partial x_j} \Big) {\sf d}x_i {\sf d}x_j.
$$

When grinding through this expression, you'll encounter product of differentials, which can be combined as

$$
{\sf d}\beta {\sf d}t = 0 \\
{\sf d}t {\sf d}\beta = 0 \\
{\sf d}\beta {\sf d}\beta^\top = Q {\sf dt},
$$

where $Q$ is the covariance matrix of the multivariate Brownian motion. The significance of the Itô formula for our purposes is that it gives us a way to get around the troublesome step in the integrating factor method. The integrating factor for our diffusion process is $\exp\{-F t\}$, which is a matrix exponential expression. Multiplying both sides by the integrating factor and rearranging, we have

$$
\exp\{-F t\} {\sf d} x - \exp\{-F t\} x {\sf d}t =
\exp\{-F t\} G u {\sf d}t + \exp\{-F t\} L w {\sf d}t.
$$

Now, we can define

$$
\phi(x, t) = \exp\{-F t\} x,
$$

and use Itô's formula to conclude that

$$
{\sf d} \big( \exp\{-F t\} x \big) =
\exp\{-F t\} {\sf d} x - \exp\{-F t\} x {\sf d}t.
$$

Plugging this back in to our rearranged expression and then integrating from $t_0$ to $t$, we see that

$$
\exp\{-F t\} x(t) - \exp\{-F t_0\} x(t_0) =
\int_{t_0}^t \exp\{-F s\} G u(s) {\sf d}s +
\int_{t_0}^t \exp\{-F s\} L w(s) {\sf d}s.
$$

By rearranging this expression, we can solve for $x(t)$

$$
x(t) =
\exp\{F(t - t_0)\} x(t_0) +
\int_{t_0}^t \exp\{F(t - s)\} G u(s) {\sf d}s +
\int_{t_0}^t \exp\{F(t - s)\} L w(s) {\sf d}s.
$$

If we know $x(t_0)$, then the only random part of the expression is the Itô integral. In the next section, we'll see how we can reason about this integral and compute the conditional density $p(x(t) \mid x(t_0))$.

Our goal in this section is to compute $p(x(t) \mid x(t_0))$.
