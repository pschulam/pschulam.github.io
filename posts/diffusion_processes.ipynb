# Diffusion Processes

This post is an introduction to the idea of a diffusion process, which is an expressive Markov model in *continuous time*. I'll describe how a diffusion process is defined and give a brief overview of the steps we take to derive probability density functions for the process.

We're going to work with the simplest type of diffusion process: linear, time homogeneous diffusion processes. These processes are described using notation that looks like a differential equation, but that has a slightly different meaning. Here is our simple diffusion process:

$$
{\sf d} x = F x {\sf d}t + G u {\sf d}t + L w {\sf d}t.
$$

The variable $x \in \mathbb{R}^p$ denotes the *state* of the process, $u \in \mathbb{R}^q$ is the *input*, and $w \in \mathbb{R}^r$ is white noise. All three of these variables are functions of time, so to be more explicit we would write $x(t)$, $u(t)$, and $w(t)$.

The first thing to note here is that the expression above *looks* like a differential equation, but it isn't because we cannot algebraically divide by ${\sf d}t$ and get something meaningful. If you're familiar with some of the basic properties of white noise, you might remember that it is nowhere differentiable, which means that $\frac{{\sf d} w}{{\sf d}t}$ does not exist. I'm glossing over details, but at a high-level this is why we can't treat the expression above like a differential equation.

## Itô Stochastic Integrals

So if it's not a differential equation, then what is it? It's a shorthand for the following integral equation:

$$
x(t) - x(t_0) =
\int_{t_0}^t F x {\sf d}t +
\int_{t_0}^t G u {\sf d}t +
\int_{t_0}^t L w {\sf d}t.
$$

The first two integrals are standard Riemann (or Lebesgue) integrals, but the third violates conditions needed to define an integral. In particular, the white noise process $w$ is unbounded and discontinuous. The Itô stochastic integral gives precise meaning to this expression.

To define the Itô integral, we'll use the idea of Brownian motion $\beta(t)$, one of the simplest stochastic processes. The key property that we'll use is that white noise can be defined as the formal derivative of Brownian motion. So we can write

$$
\frac{{\sf d} \beta}{{\sf d}t} = w
\implies
{\sf d} \beta = w {\sf d}t.
$$

In the third integral above, we can therefore make an algebraic substitution of ${\sf d}\beta$ for $w {\sf d}t$. Once we make this substitution, then we define the Itô integral to be

$$
\int_{t_0}^t L {\sf d} \beta =
\lim_{n \to \infty}
\sum_{k} L [ \beta(t_{k+1}) - \beta(t_k) ].
$$

The increment between two time points of Brownian motion is Gaussian with scale $\sigma = t_{k+1} - t_k$, so for a finite grid $t_0, \ldots, t_K$ the right-hand expression is a sum of Gaussian distributions and has a meaning. As we let the grid become increasingly fine, we get a sequence of random variables, which we can reason about rigorously using probability theory. We'll figure out the limiting random variable is shortly.

With this definition in mind, we see that pseudo-differential equation notation above describes a stochastic process where the increment $x(t) - x(t_0)$ is a random variable with distribution defined by an Itô stochastic integral.
